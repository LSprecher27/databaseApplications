In this research of the **riveting** titanic database, I am looking for someone similar to my Grandma. My grandma is 77, around 2nd class and is female (obviously). I think of my grandma as a big spender as I always bring her Amazon packages when I visit, so I will be looking for a higher fare too.

**Critera:**
* Name
* Age: 70-77
* Pclass: 2
* Sex: female
* Fare: 100 or higher


```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('titanic.db')
print("Database connected. I'm ready, Freddy!")
```

```python
query1 = """
SELECT Name, Age, Pclass, Sex, Fare
FROM Passengers
WHERE Age <= 77 AND Age >= 70 AND Pclass = 2 AND Sex = 'female' AND Fare >= 100
ORDER BY Age DESC
"""

result = pd.read_sql(query1, conn)
display(result)
```

I did not find anybody close to my Grandma with this critera, so I will now adjust it to be less specific.

**What I will Remove:**
* Fare
* Pclass

**New Critera:**
* Name
* Age
* Sex
  



```python
query2 = """
SELECT Name, Age, Pclass, Sex, Fare
FROM Passengers
WHERE Age <= 77 AND Age >= 70 AND Sex = 'female'
ORDER BY Age DESC
"""

result = pd.read_sql(query2, conn)
display(result)
```

Finally! We found one! Her name is Mrs. Tyrell William Cavendish. Her age at the time was one year younger than my grandma, she is first class (an upgrade from what I was looking for), and payed a fare of 78.85. After adjusting my critera, I was pretty surprised that I found someone this close. I'm pretty happy with my result!

```python
query3 = """
SELECT Name, Survived
FROM Passengers
WHERE Age <= 77 AND Age >= 70 AND Sex = 'female' 
ORDER BY Age DESC
"""

result = pd.read_sql(query3, conn)
display(result)
```
