```python
passenger_name = "John Smith"

# Integer
passenger_class = 3

# Float
ticket_fare = 7.25

# Boolean
survived = True

# Print everything
print("Passenger:", passenger_name)
print("Class:", passenger_class)
print("Fare: $", ticket_fare)
print("Survived:", survived)
```

```python
# Concatenation (joining strings)
first_name = "Luke"
last_name = "Sprecher"
full_name = first_name + " " + last_name
print(full_name)  # Luke Sprecher

# Repetition
line = "-" * 40
print(line)  # ----------------------------------------

# String length
name = "Luke"
print(len(name))  # 4
```

```python
# Simple print
print("Hello World")

# Multiple items (separated by commas)
print("Name:", "Luke", "Age:", 17)

# Variables
score = 97
print("Your score is:", score)

# Formatted strings (f-strings)
name = "Luke"
grade = 97
print(f"{name} earned a grade of {grade}%")
```
